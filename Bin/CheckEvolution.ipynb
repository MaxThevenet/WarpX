```python
# Import statements
import yt ; yt.funcs.mylog.setLevel(50)
import numpy as np
import scipy.constants as scc
import matplotlib.pyplot as plt
%matplotlib notebook

#ds = yt.load( './plotfiles_1MR_10CFL_main/plt{:0>5}'.format(1000) ) # Create a dataset object

def plot_it(files, step):
    ds = yt.load( files + '/plt{:0>5}'.format(step) ) # Create a dataset object
    species = 'plasma_e'
    sl = yt.SlicePlot(ds, 2, 'Ex', aspect=0.1) # Create a sliceplot object
    sl.annotate_particles(width=(10.e-6, 'm'), p_size=5, ptype=species, col='black')
    sl.annotate_particles(width=(10.e-6, 'm'), p_size=5, ptype='driver', col='orange')
    sl.annotate_grids() # Show grids
    print(files)
    print(ds.current_time)
    sl.show() # Show the plot
    
##plot_it('./plotfiles_0MR_10CFL', 2000)
plot_it('./plotfiles_0MR_05CFL', 4000)
## plot_it('./plotfiles_1MR_05CFL_main', 4000)
plot_it('./plotfiles_1MR_10CFL_main', 4000)
##plot_it('./plotfiles', 2000)
```
